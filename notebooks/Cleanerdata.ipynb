```python
#The following look for meat or type of meat within the recipes
recipes['tuna'] = recipes['ingredients'].str.count(r'[Tt]una')
#no chicken broth here
recipes['chicken'] = recipes['ingredients'].str.count(r'[Cc]hicken(?!\s[Bb]roth|\s[Ss]tock)')
recipes['shellfish'] = recipes['ingredients'].str.count(r'[Cc]rab|[Ss]hrimp|[Ll]obster|[Pp]rawn|[Oo]yster(?!\s[Mm]ushroom|\s[Cc]racker)|[Ss]callop(?!ed\s[Pp]otato)')
#looking for boolean here
recipes['meat'] = recipes['ingredients'].str.findall(r'[Tt]urkey|[fF]ish|[Bb]eef(?!\s[Bb]roth|\s[Ss]tock|\s[Ss]oup)|[Cc]hicken(?!\s[Bb]roth|\s[Ss]tock|\s[Ss]oup)|[Cc]rab|[Pp]ork|[Ss]ausage|[Hh]am|[Ll]obster|[Tt]una|[Bb]acon|[Ss]teak|[Hh]ot\s[Dd]og|[Ss]ardine|[Ss]hrimp|[Cc]horizo')
#note, I'm adding beans in there late because a lot of these recipes contain green beans and I want those to be checked for way after all of the other main protein sources
#using findall for now, will switch over to a find if it gets a little too wild.
recipes['mainprotein'] = recipes['ingredients'].str.extract(r'([Tt]urkey|[fF]ish|[Gg]round\s[Bb]eef|[Bb]eef(?!\s[Bb]roth|\s[Ss]oup|\s[Ss]tock)|[Cc]hicken(?!\s[Bb]roth|\s[Ss]oup)|[Cc]rab|[Pp]ork|[Ss]ausage|[Hh]am|[Ll]obster|[Tt]una|[Bb]acon|[Ss]teak|[Hh]ot\s[Dd]og|[Ss]ardine|[Ss]hrimp|[Cc]horizo|[Tt]ofu|[Bb]eans|[Mm]ushroom)')
#Now we're looking for starch!
recipes['pasta'] = recipes['ingredients'].str.count(r'[Pp]asta(?!\s[Ss]auce)|[Pp]enne|[Mm]acaroni|[Gg]nocchi|[Ss]paghetti|[Mm]anicotti|[Aa]ngel\s[Hh]air|[Ff]ett')
recipes['rice'] = recipes['ingredients'].str.count(r'[Rr]ice')
recipes['bread_crumbs'] = recipes['ingredients'].str.count(r'[Bb]read\s[Cc]rumb')
recipes['mainstarch'] = recipes['ingredients'].str.find(r'[Pp]asta(?!\s[Ss]auce)|[Pp]enne|[Mm]acaroni|[Gg]nocchi|[Ss]paghetti|[Mm]anicotti|[Aa]ngel\s[Hh]air|[Ff]ettuccine|[Rr]ice(?!\s[Ww]ine|e|\s[Vv]inegar)|[Bb]iscuit|[Tt]ortilla|[Cc]racker|[Bb]read|[Pp]otato|[Hh]ash\s{0,1}[Bb]rowns{0,1}')
#The following are looking for specific ingredients within the recipes, not necessarily meat
recipes['cheese'] = recipes['ingredients'].str.count(r'[Cc]heese|[Cc]heddar|[Pp]armesean|[Ss]wiss|[Mm]ozz|[Mm]onterey\s[Jj]ack|[Gg]ruyere')
#ok this next one is honestly so funny, it shows every time someone salts in the instructions. I love it.
#I'm doing this in instructions because salt is not always listed as an ingredient
recipes['salt'] = recipes['instructions'].str.count(r'[Ss]alt')
recipes['temperature'] = recipes['instructions'].str.extract(r'(\d{3}\s{0,1}.{0,2}F|\d{3}\s{0,1}.{0,2}C|\d{3}\sdegrees|\d{3})')
#Let's get some wild ones in here
#I'm using contains to get a boolean. I might try using that for the other ones. 
#Do I think I can get away with just adding a lightning bolt button on there to filter for microwaved or non-microwaved?
recipes['microwave'] = recipes['instructions'].str.contains(r'[Mm]icrowave')
recipes['crockpot'] = recipes['instructions'].str.count(r'[Ss]low\s[Cc]ooker')
recipes['ingredient_count'] = recipes['ingredients'].str.count(r'\*{3}')+1
#Baking dish size next, this one is going to be a doozie. Keep looking into it because you might not be able to do quotation marks Do I need to denote the period with a \ ?
recipes['baking_dish'] = recipes['instructions'].str.extract(r'(\d{1,2}.{0,7}\d{1,2}["\.124/\s]{0,4}|\d{1,2}[/124"xbyin\.\s-]{0,7}\d{1,2}["\.124/\s]{0,4}|\d{1,2}.[Ii]nch.\d{1,2}.["\.124/\s]{0,4}[Ii]nch|\d{1,2}.{0,1}[Ii]nch|\d{1,2}[\.\s124/]{0,5}[Qq]uart|\d{1,2}.[Qq]t.)')
#this might simplify it by a lot. Will need some fanangling.
recipes['baking_dish_simple'] = recipes['instructions'].str.extract(r'(\d.{0,30}[Dd]ish|\d.{0,30}killet|\d.{0,30}[Pp]an)')
recipes['dish_type'] = recipes['instructions'].str.findall(r'[Gg]lass|[Cc]eramic|[Cc]ast\s[Ii]ron|[Dd]utch\s[Oo]ven')
recipes
```
