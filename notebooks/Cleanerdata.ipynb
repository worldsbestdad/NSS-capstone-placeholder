```python
recipes = pd.DataFrame()
recipes['titles'] = pd.concat([bc['titles'], ar['titles'], ba['titles'], f['titles'], sl['titles'], pw['titles']])
recipes['rating'] = pd.concat([bc['rating'].astype('str'), ar['rating'].astype('str'), ba['rating'].astype('str'), f['rating'].astype('str'), sl['rating'].astype('str'), pw['rating'].astype('str')])
recipes['ingredients'] = pd.concat([bc['ingredients'], ar['ingredients'], ba['ingredients'], f['ingredients'], sl['ingredients'], pw['ingredients']])
recipes['instructions'] = pd.concat([bc['instructions'], ar['instructions'], ba['instructions'], f['instructions'], sl['instructions'], pw['instructions']])
recipes['total_time'] = pd.concat([bc['total_time'], ar['total_time'], ba['total_time'], f['total_time'], sl['total_time'], pw['total_time']])
recipes['host'] = pd.concat([bc['host'], ar['host'], ba['host'], f['host'], sl['host'], pw['host']])
#recipes['nutrients'] = pd.concat([bc['nutrients'], ar['nutrients'], ba['nutrients'], f['nutrients'], sl['nutrients'], pw['nutrients']])
recipes
```

```python
#The following look for meat or type of meat within the recipes
recipes['tuna'] = recipes['ingredients'].str.count(r'tuna|Tuna')
#no chicken broth here
recipes['chicken'] = recipes['ingredients'].str.count(r'chicken\s(?!broth|Broth)|Chicken\s(?!broth|Broth)')
recipes['crab'] = recipes['ingredients'].str.count(r'crab|Crab')
recipes['meat'] = recipes['ingredients'].str.count(r'fish|Fish|beef|Beef|chicken\s(?!broth|Broth)|Chicken\s(?!broth|Broth)|crab|Crab|pork|Pork|sausage|Sausage|ham|Ham|lobster|Lobster|tuna|Tuna|bacon|Bacon|steak|Steak|hot\sdog|Hot\sdog')
#Now we're looking for starch!
recipes['pasta'] = recipes['ingredients'].str.count(r'pasta\s(?!sauce|Sauce)|Pasta\s(?!sauce|Sauce)|penne|Penne|macaroni|Macaroni')
recipes['rice'] = recipes['ingredients'].str.count(r'rice|Rice')
recipes['bread_crumbs'] = recipes['ingredients'].str.count(r'bread\scrumb|Bread\scrumb|Bread\sCrumb')
#The following are looking for specific ingredients within the recipes, not necessarily meat
recipes['cheese'] = recipes['ingredients'].str.count(r'cheese|cheese|cheddar|Cheddar|parmesean|Parmesean')
#ok this next one is honestly so funny, it shows every time someone salts in the instructions. I love it.
recipes['salt'] = recipes['instructions'].str.count(r'salt|Salt')
recipes['temperature'] = recipes['instructions'].str.findall(r'\d{3}\s{0,1}.F|\d{3}\s{0,1}.C|\d{3}\sdegrees')
#Let's get some wild ones in here
recipes['microwave'] = recipes['instructions'].str.count(r'microwave|Microwave')
recipes['crockpot'] = recipes['instructions'].str.count(r'slow\scooker|Slow\scooker|Slow\sCooker')
recipes['ingredient_count'] = recipes['ingredients'].str.count(r'\*{3}')+1
#Baking dish size next, this one is going to be a doozie.
recipes['baking_dish'] = recipes['instructions'].str.find(r'baking\sdish|Baking\sdish')
recipes['dish_type'] = recipes['instructions'].str.findall(r'glass|Glass|ceramic|Ceramic|cast\siron|Cast\sIron|Cast\siron')
```

```python
recipes = recipes.drop(recipes[recipes.titles == 'None'].index)
```

```python
recipes.to_excel("../data/casseroles.xlsx")
```
