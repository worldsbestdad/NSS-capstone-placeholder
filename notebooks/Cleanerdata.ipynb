```python
recipes = pd.DataFrame()
recipes['titles'] = pd.concat([bc['titles'], ar['titles'], ba['titles'], f['titles'], sl['titles'], pw['titles']])
recipes['rating'] = pd.concat([bc['rating'].astype('str'), ar['rating'].astype('str'), ba['rating'].astype('str'), f['rating'].astype('str'), sl['rating'].astype('str'), pw['rating'].astype('str')])
recipes['ingredients'] = pd.concat([bc['ingredients'], ar['ingredients'], ba['ingredients'], f['ingredients'], sl['ingredients'], pw['ingredients']])
recipes['instructions'] = pd.concat([bc['instructions'], ar['instructions'], ba['instructions'], f['instructions'], sl['instructions'], pw['instructions']])
recipes['total_time'] = pd.concat([bc['total_time'], ar['total_time'], ba['total_time'], f['total_time'], sl['total_time'], pw['total_time']])
recipes['host'] = pd.concat([bc['host'], ar['host'], ba['host'], f['host'], sl['host'], pw['host']])
#recipes['nutrients'] = pd.concat([bc['nutrients'], ar['nutrients'], ba['nutrients'], f['nutrients'], sl['nutrients'], pw['nutrients']])
recipes
```

```python
tuna_regex = 'tuna'
recipes['tuna'] = recipes['ingredients'].str.findall('tuna')
recipes['chicken'] = recipes['ingredients'].str.findall('chicken')
recipes['crab'] = recipes['ingredients'].str.findall('crab')
#ok this next one is honestly so funny, it shows every time someone salts in the instructions. I love it.
recipes['salt'] = recipes['instructions'].str.count('salt')
recipes['temperature'] = recipes['instructions'].str.findall(r'\d{3}°F')
```

```python
recipes.to_excel("../data/casseroles.xlsx")
```
